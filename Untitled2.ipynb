```python
import matplotlib.pyplot as m
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
m.plot(x,y)
m.show()
```

```python
m.figure(figsize=(4,3)) # figure size
m.plot(x, y, color='blue', marker='o', linestyle='--', linewidth=2, markersize=12)
m.title("Kuch bhi Title hai")
m.xlabel("x-axis label hai")
m.ylabel("y-axis label hai")
m.show()
```

```python
#data
x = [1,2,3,4,5]
y1 = [10, 20, 25, 35, 45]
y2 = [20, 30, 35, 45, 55]
m.plot(x, y1, label='Sales 2024') # ploting y1 data
m.plot(x, y2, label='Sales 2025') # ploting y2 data
m.title("YoY Sales")
m.xlabel("Months")
m.ylabel("Sales")
m.legend() # show legends
m.show()
```
